```python
import matplotlib.pyplot as plt
import numpy as np
import os

from scipy import misc
from joblib import dump, load

import theano.tensor as T
import lasagne.layers
import theano
from collections import OrderedDict

from nn_utils import build_UNet

%run processing_functions.ipynb
import copy

import time 
from IPython.display import clear_output

import joblib
%matplotlib inline
```

```python
def get_data(dir_path):
    X = []
    for i, fname in enumerate(os.listdir(dir_path)):
        img_X = misc.imread(dir_path + fname)
        img_X = (img_X.astype(np.float32) / 255 - 0.5)
        img_X = img_X[:, :, np.newaxis].transpose(2, 0, 1)
        X.append(img_X)
    return X
```

```python
X = get_data("test/")
```

```python
patch_size = 128
image_h = 1000
image_w = 1000
base_n_filters = 32
```

```python
input_image = T.tensor4('input')
target = T.matrix('output', dtype='int64')

net = build_UNet(n_input_channels=1, BATCH_SIZE=None, num_output_classes=2,
                 pad='same', nonlinearity=lasagne.nonlinearities.elu,
                 input_dim=(patch_size, patch_size), base_n_filters=base_n_filters, do_dropout=True,
                 weights=load("weights/UNet_params_ep009.jbl"))
predictions = lasagne.layers.get_output(net['output_flattened'], input_image)
get_predictions = theano.function([input_image], predictions)
```

```python
rough_filter_size = 11
accurate_filter_size = 11
rough_threshold = 0.1
accurate_threshold = 0.99
```

```python
rough_detector = build_background_detector(None, (image_h, image_w), rough_filter_size, rough_threshold)
accurate_detector = build_background_detector(None, (image_h, image_w), accurate_filter_size, accurate_threshold)

output_rough_background = lasagne.layers.get_output(rough_detector['background_detector'], input_image)
output_accurate_background = lasagne.layers.get_output(accurate_detector['background_detector'], input_image)

get_rough_background = theano.function([input_image], output_rough_background)
get_accurate_background = theano.function([input_image], output_accurate_background)
```

```python
binarization_threshold = 0.3
closing_filter_size = 8
minimal_crack_size = 100
```

```python
start_time = time.time()
image = X[3][0]
processed_image, background, bins, weights = process_image(image, get_predictions, get_rough_background, get_accurate_background, 
                                patch_size=128, n_input_chanels=1,
                                binarization_threshold=binarization_threshold,
                                closing_filter_size=closing_filter_size, 
                                minimal_crack_size=minimal_crack_size)
print time.time() - start_time
```

```python
dir_path = "3/"
```

```python
stack_size = len(os.listdir(dir_path))
```

```python
horizontal = np.zeros((stack_size, image_w))
vertical = np.zeros((stack_size, image_h))
```

```python
fnames = sorted(os.listdir(dir_path), key=lambda x: int(x[:5]))
```

```python
cum_weights = np.zeros(37)
```

```python
for i, fname in enumerate(fnames):
    image = misc.imread(dir_path + fname)
    image = (image.astype(np.float32) / 255 - 0.5)
    #image = image[:, :, np.newaxis].transpose(2, 0, 1)
    processed_image, background, bins, weights = process_image(image, get_predictions, get_rough_background, get_accurate_background, 
                                patch_size=128, n_input_chanels=1,
                                binarization_threshold=binarization_threshold,
                                closing_filter_size=closing_filter_size, 
                                minimal_crack_size=minimal_crack_size)
    cum_weights += weights
    background = background.reshape((image_h, image_w))
    
    horizontal[i] = np.where(np.logical_and(background[image_h // 2, :] == 1, processed_image[image_h // 2, :] == 0),
                             0.4, processed_image[image_h // 2, :])
    vertical[i] = np.where(np.logical_and(background[:, image_w // 2] == 1, processed_image[:, image_w // 2] == 0),
                             0.4, processed_image[:, image_w // 2])
    
    if i % 50 == 0:
        clear_output()
        print i
        plt.figure(figsize=(14, 10))
        plt.imshow(horizontal, cmap='gray')
        plt.figure(figsize=(14, 10))
        plt.imshow(vertical, cmap='gray')
        plt.show()
    misc.imsave('3_ans/ans_' + fname, np.where(np.logical_and(background == 1, processed_image == 0),
                             0.4, processed_image))
    #plt.figure(figsize=(16, 8))
    #plt.subplot(1, 2, 1)
    #plt.imshow(image, cmap='gray')
    #plt.subplot(1, 2, 2)
    #plt.imshow(np.where(np.logical_and(accurate_background[0, 0] == 1, processed_image == 0), 0.4, processed_image),
    #           cmap='gray')
    #plt.savefig('examples/ex' + str(i) + '.png')
```
